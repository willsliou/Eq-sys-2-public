```python
import os
import subprocess
import time
import pandas as pd

npx_path = r'"C:\Program Files\nodejs\npx.cmd"'
node_path = r'"C:\Program Files\nodejs"'
npx_path_duka = r'"C:\Users\hello\AppData\Roaming\npm\dukascopy-node.cmd"'


# Update the PATH environment variable
os.environ["PATH"] += os.pathsep + node_path

def file_exists(directory, symbol):
    """
    Check if a file exists in the given directory with the given symbol name.
    """
    for local_file_name in os.listdir(directory):
        if local_file_name.endswith('.json'):
            # Get id
            ticker = local_file_name.split("-")[0]

            # Compare ids
#             print(f"Comparing {ticker} | {symbol}")
            if ticker == symbol:
                print(f"Match found in directory: {ticker} == {symbol}")
                return True
    return False

def run_subprocess(cmd, symbol):
    """
    Run the given subprocess command and print the output.
    """
    try:
        # Set the PATH environment variable for the subprocess
        env = os.environ.copy()
        print(f'Executing on {symbol}')
        process = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, env=env)
        stdout, stderr = process.communicate()
        print(stdout)

        if process.returncode != 0:
            raise subprocess.CalledProcessError(process.returncode, cmd, output=stdout, stderr=stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error while processing {symbol}: {str(e)}")
        print("Output:", e.output)
        print("Error output:", e.stderr)

def sleep(interval):
    """
    Sleep for the given interval in seconds and print a message.
    """
    print(f"Sleeping for {interval} seconds...")
    time.sleep(interval)
    print("Resuming...")
    
    
def periodic_work(interval, list_urls):
    while True:
        print("Starting...")
        # Collect symbols to execute
        collect_symb = []    
        for url in list_urls:
            table_id = pd.read_html(url)[0]['id']
            directory = 'C:/Users/hello/Documents/GitHub/eq-sys-1/download'        
            for symbol in table_id:
#                 print("Checking if file exists...")
                if file_exists(directory, symbol):
                    print("Table already exists. Checking for a new symbol...")
                else:
                    print(f"Appending {symbol}")
                    collect_symb.append(symbol)

        # Execute commands
        commands = []
        processes = []
        while True:
            # Add new commands as subprocesses finish
            while len(commands) > 0 and len(processes) < 100:
                cmd = commands.pop(0)
                processes.append(subprocess.Popen(cmd, shell=True))

            # Check which subprocesses have finished
            for process in processes:
                if process.poll() is not None:
                    # Process has finished
                    processes.remove(process)
                    stdout, stderr = process.communicate()
#                     print(stderr.decode())
# print*
                    if process.returncode != 0:
                        print(f"Error while processing {process.args}: returncode={process.returncode}")
                        print("Error output:", stderr.decode())
                    else:
                        print(f"Finished processing {process.args}")
                        
                        # Add a new command to execute
                        if len(collect_symb) > 0:
                            symbol = collect_symb.pop(0)
                            cmd = f'{npx_path} {npx_path_duka} -i {symbol} -from 1980-06-01 -to 2023-04-09 -t m1 -bs 15 -bp 2000'
                            commands.append(cmd)
                            print(f"Added new command to execute: {cmd}")

            # Add new commands to execute
            while len(commands) < 100 and len(collect_symb) > 0:
                symbol = collect_symb.pop(0)
                cmd = f'{npx_path} {npx_path_duka} -i {symbol} -from 1980-06-01 -to 2023-04-09 -t m1 -bs 15 -bp 2000'
                commands.append(cmd)
                print(f"Added new command to execute: {cmd}")
                
            # Check if all processes are finished
            if len(processes) == 0 and len(collect_symb) == 0:
                print("All commands executed. Exiting...")
                break
                
            # Sleep for a short interval
            time.sleep(1)
```
