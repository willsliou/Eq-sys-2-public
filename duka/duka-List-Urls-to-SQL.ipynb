```python
# Create the SQL command to create a new database
engine = create_engine('sqlite:///Duka_LIST_URLS_ITEMS.db')

# Loop over the list of table names and create a new table for each one

# Loop of both tables lst and list_urls. Get name from list_urls. Append corresponding table from lst
for i in range(len(lst)):
    table_name = list_urls[i].split('/')[-1]
    print(table_name)
    table_to_append = lst[i]
    try:
        # Execute the SQL command to create the new table
        table_to_append.to_sql(table_name , engine, if_exists='fail')
        print(f"Table '{table_name}' created successfully")

    except Exception as e:
        print(f"Table {table_name} already exists. Error: ", e)
    
# Close the database connection
engine.dispose()
```
