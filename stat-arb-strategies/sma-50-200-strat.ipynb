```python
total_years = []
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
df = pd.DataFrame(arr, columns=['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12'], index=['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12'])

starting_year = 2011
ending_year = 2022
for i in range(starting_year, ending_year):
    for mnth in range(12):
        for mnth_2 in range(mnth, 12):
            starting_month = months[mnth]
            ending_month = months[mnth_2]
            new_beg = str(i) + '-' + str(starting_month) + '-' + "01"
            # print(new_beg)
            new_end = str(i) + '-' + str(ending_month) + '-' + "28"
            # print(str(months[mnth_2]))
            data = yf.download("AMZN", start = new_beg, end=new_end)
            
            starting_price = data['Adj Close'][0]
            ending_price = data['Adj Close'][-1]
            price_diff = ending_price - starting_price
            
            df[starting_month][ending_month] += price_diff # Append price to corresponding Starting Month v Ending Month in matrix

            total_years.append((new_beg + ' to ' + new_end, price_diff))

total_years
```

```python
num_years = ending_year-starting_year
for i in range(12):
    for j in range(12):
        starting_month = months[i]
        ending_month = months[j]
        df[starting_month][ending_month] /= num_years
df
```
